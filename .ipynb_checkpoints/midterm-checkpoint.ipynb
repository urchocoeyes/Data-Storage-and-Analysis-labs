# ANOVA
   1.H0: When all samples’ means are the same.
   2.H1: When one or more samples are very much different.


```python
from scipy.stats import f_oneway
class1 = df['weight_kg']
class2 = df['body fat_%']
class3 = df['sit-ups counts'] 
print(f_oneway(class1, class2, class3))
```

F-statistic = 50205.49 → A very high F-value suggests strong variance between at least one of the groups.
p-value = 0.0 (practically 0) → This is extremely small, meaning the probability of the null hypothesis being true is almost 0.


```python
#Boxlot can plot outliers in data
plt.figure(figsize = (20, 10))
sns.boxplot(data = df)
```

```python
# removing outliers in 'systolic' column:
sns.boxplot(df['systolic'])
plt.title('Boxplot before removing Outliers')
plt.show()
        
```

```python
#removing Outliers
for i in df['systolic']:
    q1 = df['systolic'].quantile(0.25)
    q3 = df['systolic'].quantile(0.75)
    iqr = q3-q1
    lower_tail = q1 - 1.5*iqr
    upper_tail = q3 + 1.5*iqr
    
    if i>upper_tail or i<lower_tail:
        df['systolic'] = df['systolic'].replace(i,0)
        
sns.boxplot(df['systolic'])
plt.title('Boxplot after removing Outliers')
plt.show()
```
