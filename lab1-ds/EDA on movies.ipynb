```python
def convert_gross_value(value):
    if pd.isna(value):
        return 0
    value = value.replace(',', '').replace('$', '')
    if 'M' in value:
        value = value.replace('M', '')
        value = float(value) * 1_000_000  # Convert from M to actual number
    elif 'K' in value:
        value = value.replace('K', '')
        value = float(value) * 1_000  # Convert from K to actual number
    else:
        value = float(value)  # Convert string to float directly
    return value

df['Gross'] = df['Gross'].apply(convert_gross_value)
```

```python
df['Gross']
```

```python
df['MOVIES'] = df['MOVIES'].str.strip()
df['GENRE'] = df['GENRE'].str.strip()
df['ONE-LINE'] = df['ONE-LINE'].str.strip()
df['STARS'] = df['STARS'].str.strip()
df['ONE-LINE'] = df['ONE-LINE'].replace('Add a Plot', 'Unknown')
df['ONE-LINE'] = df['ONE-LINE'].str.replace('[^\w\s]', '', regex=True)
```

```python
def extract_information(text):
    director = None
    stars = None
    if 'Director:' in text:
        director = text.split('Director:\n')[1].split('\n|')[0].strip()
    if 'Stars:' in text:
        stars = text.split('Stars:\n')[1].strip()
    return director, stars

df[['Director', 'Stars']] = df['STARS'].apply(lambda x: pd.Series(extract_information(x)))

df.drop('STARS', axis=1, inplace=True)
```

```python
df['Stars'] = df['Stars'].apply(lambda x: ' '.join(x.split()) if x and x != 'Unknown' else 'Unknown')
```

```python
df.isna().sum()
```

```python
def fill_with_median(series):
    if series.dropna().empty:
        return series
    else:
        return series.fillna(series.median())
```

```python
df['YEAR'] = df.groupby('GENRE')['YEAR'].transform(fill_with_median)
df['GENRE'] = df['GENRE'].fillna(df['GENRE'].mode()[0])
df['RunTime'] = df.groupby(['GENRE', 'RATING'])['RunTime'].transform(fill_with_median)
df['Director'] = df['Director'].fillna('Unknown')
df['RATING'] = df.groupby(['GENRE', 'YEAR'])['RATING'].transform(fill_with_median)
df['VOTES'] = df.groupby(['GENRE', 'YEAR'])['VOTES'].transform(fill_with_median)
```

```python
df.isna().sum()
```

```python
df['RunTime'] = df['RunTime'].fillna(df['RunTime'].median())
df['YEAR'] = df['YEAR'].fillna(df['YEAR'].median())
df['RATING'] = df['RATING'].fillna(df['RATING'].median())
df['VOTES'] = df['VOTES'].fillna(df['VOTES'].median())
```

```python
df.isna().sum()
```

```python
df['YEAR'] = df['YEAR'].astype(int)
df['RunTime'] = df['RunTime'].astype(int)
df['VOTES'] = df['VOTES'].astype(int)

df.info()
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR
    
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

cols = ['VOTES', 'RunTime']

for column in cols:
    df = remove_outliers(df, column)
df
```

## Key Patterns and Trends:
* YEAR vs. Other Attributes:
There is a very weak positive correlation between "YEAR" and "RATING" (0.02), suggesting that the release year of movies doesn't significantly affect their ratings.


* RATING vs. Other Attributes:
"RATING" has a moderate negative correlation with "RunTime" (-0.36), which suggests that movies with lower ratings tend to have longer runtimes.
"RATING" shows a weak positive correlation with "VOTES" (0.19), implying that movies with higher ratings tend to receive more votes.

* VOTES vs. Other Attributes:
"VOTES" shows a slight positive correlation with "RunTime" (0.19), suggesting that longer movies might have slightly more votes.
There is a weak correlation between "VOTES" and "Gross" (0.1), indicating a minor relationship between the number of votes and the movie's box office earnings.


## Recommendations:
Marketing and Audience Engagement:
Movies with higher ratings tend to attract more votes. Therefore, enhancing the quality of movies to boost ratings may lead to increased audience engagement and more votes.
Optimizing Movie Length:
Since longer movies have a moderate negative correlation with ratings, studios might consider focusing on shorter films to improve viewer satisfaction and ratings.
